```python
# Получим заголовки
if len(rows) > 0:
  headers = "<tr>"
  for key in rows[0]:
    headers += "<th>"+key+"</th>\n"
  headers += "</tr>"

# Выведем по очереди все строки и столбцы в них
table = ''
for row in rows:
  table += "<tr>"
  for col in row:
    table += "<td>"+str(row[col])+"</td>"
  table += "</tr>"


display(HTML("<h1>Билеты</h1><table border=1 cellpadding=10>"+headers+table+"</table>"))
```
