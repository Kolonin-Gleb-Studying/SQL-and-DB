```python
def insertDB(dbh, rnd1, rnd2):
  try:
    with dbh.cursor() as cur:
      sql = "INSERT INTO random (rnd1, rnd2) \
       VALUES('"+str(rnd1)+"', "+str(rnd2)+")"
      print(sql)
      cur.execute(sql)
  except:
    print("Что-то пошло не так")
```

```python
for i in range(10):
  rnd1 = random.uniform(0, 1)
  rnd2 = random.randint(-1000, 1000)
  insertDB(dbh, rnd1, rnd2)
  #print(rnd1, rnd2)
dbh.commit()
```

```python
try:
  with dbh.cursor() as cur:
    cur.execute('SELECT * FROM random')
    rows = cur.fetchall()
    print(rows)
except:
  print("Что-то пошло не так")
```

```python
headers = ''
# Получим заголовки
if len(rows) > 0:
  headers = "<tr>"
  for key in rows[0]:
    headers += "<th>"+key+"</th>\n"
  headers += "</tr>"

# Выведем по очереди все строки и столбцы в них
table = ''
for row in rows:
  table += "<tr>"
  for col in row:
    table += "<td>"+str(row[col])+"</td>"
  table += "</tr>"


display(HTML("<h1>Случайные числа</h1><table border=1 cellpadding=10>"+headers+table+"</table>"))
```
